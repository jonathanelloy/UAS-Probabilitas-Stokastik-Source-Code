```python
def clean(tahun, bundir, populasi):
    out = [[],[],[]]
    for i in range(len(tahun)):
        if tahun[i] not in out[0]:
            out[0].append(tahun[i])
            out[1].append(bundir[i])
            out[2].append(populasi[i])
        else:
            out[1][out[0].index(tahun[i])]+=bundir[i]
            out[2][out[0].index(tahun[i])]+=populasi[i]
    for i in range(len(out[0])-1):
        for j in range(len(out[0])-1):
            if out[0][j]>out[0][j+1]:
                temp = out[0][j]
                temp_2 = out[1][j]
                temp_3 = out[2][j]
                out[0][j] = out[0][j+1]
                out[0][j+1] = temp
                out[1][j] = out[1][j+1]
                out[1][j+1] = temp_2
                out[2][j] = out[2][j+1]
                out[2][j+1] = temp_3
    return out
            

a = clean(data_fix['Tahun'],data_fix['Banyak Bunuh Diri'], data_fix['Populasi'])
a
```

```python
n=np.size(a[0][0:-1])
year = np.array(a[0][0:-1])
count = np.array([(a[1][i]/a[2][i])*100000 for i in range(len(a[1])-1)])
def estimate_coef(x, y): 
    # number of observations/points 
    n = np.size(x) 
  
    # mean of x and y vector 
    m_x= np.mean(x) 
    m_y= np.mean(y) 
  
    # calculating cross-deviation and deviation about x 
    SS_xy = np.sum(y*x) - n*m_y*m_x 
    SS_xx = np.sum(x*x) - n*m_x*m_x 
  
    # calculating regression coefficients 
    b_1 = SS_xy / SS_xx 
    b_0 = m_y - b_1*m_x 
  
    return(b_0, b_1) 
def plot_regression_line(x, y, b): 
    plt.figure(figsize=(20,10))
    # plotting the actual points as scatter plot 
    plt.scatter(x, y, color = "m", 
               marker = "o", s = 30) 
  
    # predicted response vector 
    y_pred = b[0] + b[1]*x 
  
    # plotting the regression line 
    plt.plot(x, y_pred, color = "g") 
    
    plt.grid(b=None,which='major',axis='both')
    # putting labels 
    plt.xlabel('Tahun') 
    plt.ylabel('Bunuh Diir/100k Penduduk') 
  
    # function to show plot 
    plt.show() 

b = estimate_coef(year, count)
plot_regression_line(year,count,b)
```

```python
print("\nAuto Correlation Plot")
autocorrelation_plot(count)
model = ARIMA(count, order=(5,1,0))
model_fit = model.fit(disp=0)
print(model_fit.summary())
residuals = DataFrame(model_fit.resid)
residuals.plot()
plt.show()
residuals.plot(kind='kde')
plt.show()
print(residuals.describe())

```

```python
#from sklearn.metrics import MeanSquaredError
def StartARIMAForecasting(Actual, P, D, Q):
    model = ARIMA(Actual, order=(P, D, Q))
    model_fit = model.fit(disp=0)
    prediction = model_fit.forecast()[0]
    return prediction

NumberOfElements = len(count)
TrainingSize = int(NumberOfElements * 0.7)
TrainingData = count[0:TrainingSize]
TestData = count[TrainingSize:NumberOfElements]
Actual = [x for x in TrainingData]
Predictions = []
for timepoint in range(len(TestData)):
    ActualValue =  TestData[timepoint]
    #forcast value
    Prediction = StartARIMAForecasting(Actual, 5,1,0)    
    print('Actual=%f, Predicted=%f' % (ActualValue, Prediction))
    #add it in the list
    Predictions.append(Prediction)
    Actual.append(ActualValue)

plt.plot(TestData, label='Aktual')
plt.plot(Predictions, color='red',label='Prediksi')
plt.legend()S
plt.show()

```
